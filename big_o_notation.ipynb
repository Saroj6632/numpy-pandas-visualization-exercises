```python
import matplotlib.pyplot as plt
import math
```

```python



x = range(1,20)
y = [1 for n in x]
y1 = [math.log(n) for n in x]
y2 = [n for n in x]
y3 = [n * math.log(n) for n in x]
y4 = [n**2 for n in x]
y5 = [2**n for n in x]
y6 = [math.factorial(n) for n in x]
y7 = [n**n for n in x]

plt.figure(figsize = (12,8))
plt.plot(x, y, label='$\mathcal{O}(1)$')
plt.plot(x, y1, label='\$mathcal{O}(logn)$')
plt.plot(x, y2, label='\$mathcal{O}(n)$')
plt.plot(x, y3, label='$\mathcal{O}(nlog n)$')
plt.plot(x, y4, label='$\mathcal{O}(n^{2})$')
plt.plot(x, y5, label='$\mathcal{O}(2^{n})$')
plt.plot(x, y6, label='$\mathcal{O}(n!)$')
plt.plot(x, y7, label ='$\mathcal{O}(n^{n})$')


plt.title('Bog O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='upper right')
plt.ylim(0,50)




```
