```python
# Load the mpg dataset from PyDataset.


from pydataset import data
mpg = data('mpg')
mpg.head()

```

```python
# Check out your column names and perform any cleanup you may want on them.


mpg = mpg.rename(columns={'cty': 'city','hwy':'highway'})
mpg
```

```python
# Display the summary statistics for the dataset.

mpg.describe()
```

```python
# How many different manufacturers are there?

len(mpg.manufacturer.value_counts())
```

```python
# Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg.head()
```

```python
# Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

mpg['average_mileage'] = (mpg.city + mpg.highway) / 2
mpg.head()
        
```

```python
# Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

mpg['is_automatic'] = np.where(mpg.trans.str.startswith('auto'),'Auto', 'Manual')
mpg.head(10)
```

```python
# Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?
 # Honda has best miles per gallon

mpg.groupby('manufacturer').average_mileage.agg(['mean']).sort_values(by='mean',ascending = False)


```

```python
# Do automatic or manual cars have better miles per gallon?
# manual has better miles per gallon
mpg.groupby('is_automatic')[['average_mileage']].mean()
```

```python
# Use your get_db_url function to help you explore the data from the chipotle database.

chipotle = pd.read_sql('SELECT * FROM orders', get_db_url(username,password,host,database_name ='chipotle'))
chipotle.head(20)
```

```python
chipotle.info()
```

```python
# What is the total price for each order?
 # first change the column item_price to float type as it is object
dollarize_price = (chipotle.item_price.apply(lambda x: float(x[1:-1])))
dollarize_price






```

```python
chipotle['total_price'] = chipotle.quantity * dollarize_price
chipotle.head()
```

```python
# What are the most popular 3 items?
chipotle.groupby(by="item_name").sum().sort_values('quantity',ascending=False).head(3)


```

```python
# Which item has produced the most revenue?
#chicken bowl has highest revenue
chipotle.groupby(by="item_name").sum().sort_values('total_price',ascending=False).head()
## or using nlargest method
# chipotle.groupby(by="item_name").sum().sort_values('total_price',ascending=False).nlargest(n=1, 'total_price')


```

```python
# Join the employees and titles DataFrames together.

employees = pd.read_sql('SELECT * FROM employees', get_db_url(username,password,host,database_name='employees'))
employees.head()


```

```python
titles = pd.read_sql('SELECT * FROM titles', get_db_url(username,password,host,database_name='employees'))

titles.head()
```

```python
new_table = employees.merge(titles, left_on='emp_no',right_on='emp_no', how='outer',indicator=True)

new_table.head()
```

```python
# For each title, find the hire date of the employee that was hired most recently with that title.

new_table.groupby(by='title').from_date.max()
```
