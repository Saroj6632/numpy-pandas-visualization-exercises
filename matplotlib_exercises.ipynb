```python
## Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y=root x
plt.figure(figsize=(5,5))
import math
x = range(0,5)
y = [math.sqrt(x) for x in x]
plt.plot(x,y)

```

```python
#y = x^3
plt.figure(figsize=(3,5))
x = range(0,5)
y = [math.pow(x, 3) for x in x]
plt.plot(x,y)
```

```python
# y =2^x
plt.figure(figsize=(3,5))
x = range(0,5)
y = [(2 ** x) for x in x]
plt.plot(x,y)
```

```python
# y =1/(x+ 1)
plt.figure(figsize=(3,5))
x = range(0,5)
y = [(1/(x + 1)) for x in x]
plt.plot(x,y)
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize=(8,6))
x = range(0,5)
y = [math.sqrt(x) for x in x]
z = [(2 ** x) for x in x]
v = [(1/(x + 1)) for x in x]
s = [math.pow(x, 3) for x in x]

#first plot
plt.subplot(221)
plt.plot(x,y)
plt.title('square Root of x')

# 2nd plot
plt.subplot(222)
plt.plot(x,z)
plt.title('2 power x')

#3rd plot
plt.subplot(223)
plt.plot(x,v)
plt.title('some func x')

#4th plot
plt.subplot(224)
plt.plot(x,s)
plt.title('Cube of x')

plt.suptitle('subplots')



```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an a
#appropriate title for the figure.
plt.figure(figsize=(10,10))
x = range(0,5)
y = [math.sqrt(x) for x in x]
z = [(2 ** x) for x in x]
v = [(1/(x + 1)) for x in x]
s = [math.pow(x, 3) for x in x]

plt.plot(x,y, c='red', label='SquareRoot')
plt.plot(x,z, c='green',label ='2^x')
plt.plot(x,v, c='black',label='simple math')
plt.plot(x,s, c='blue', label='cube')

plt.legend(loc='upper left')

plt.title('Maths expressions')


```
